# CLI Design Specification: `quantum_qr`

## Usage Examples
* **Generate (Phase 3):** `python -m quantum_qr generate "pay alice $10" --output data/alice.png --bits 8`
* **Verify (Phase 4):** `python -m quantum_qr verify data/alice.png`

## Subcommands
* `generate`: Creates a tamper-evident QR code from payload data.
* `verify`: (Reserved for Phase 4) Reads a QR code and runs the quantum Deutsch-Jozsa verifier against the expected secret.

## `generate` Arguments
| Argument | Type | Required | Default | Description |
| :--- | :--- | :---: | :---: | :--- |
| `data` | Positional | Yes | - | The string payload to encode (e.g., `"pay alice $10"`). |
| `--output`, `-o` | Option | Yes | - | Filepath to save the generated QR image. |
| `--bits`, `-n` | Option | No | `8` | Number of bits for the HMAC truncation and quantum register. |
| `--json` | Flag | No | `False` | Outputs machine-readable JSON to stdout instead of human-friendly text. |

## Exit Codes
| Code | Meaning | Description |
| :---: | :--- | :--- |
| `0` | **Success** | Execution completed flawlessly. |
| `1` | **App Error** | Runtime exception (e.g., payload too large, invalid data, missing keys). |
| `2` | **Usage Error** | Invalid arguments or missing required flags (handled automatically by `argparse`). |
