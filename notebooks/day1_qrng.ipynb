```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
# Create a Quantum Circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply a Hadamard gate to qubit 0 to create a superposition
qc.h(0)

# Measure qubit 0 into classical bit 0
qc.measure(0, 0)

# Optional: Draw the circuit to visually verify it
qc.draw('mpl')
```

```python
# Initialize the local Aer simulator backend
simulator = AerSimulator()

# Run the circuit once to get a single random shot
job = simulator.run(qc, shots=1)
result = job.result()

# Extract the counts dictionary (e.g., {'0': 1} or {'1': 1})
counts = result.get_counts()
print("Measurement counts:", counts)
```

```python
# Extract the single bit string key from the dictionary
random_bit = list(counts.keys())[0]

print(f"Generated Random Bit: {random_bit}")
```

```python
from qiskit.visualization import plot_histogram

# Run the same circuit with 1024 shots to collect statistical data
job_stat = simulator.run(qc, shots=1024)
result_stat = job_stat.result()

# Extract the distribution counts
counts_stat = result_stat.get_counts()
print("Statistical Counts (1024 shots):", counts_stat)

# Plot and display the histogram distribution
plot_histogram(counts_stat)
```
